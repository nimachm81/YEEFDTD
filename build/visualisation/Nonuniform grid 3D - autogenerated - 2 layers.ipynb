```python
## animate Electric field

%pylab tk
import numpy as np
from matplotlib import pyplot as plt
from ReadBinary import *

indStart = 0

Ex_m_xplane = GetArrays("../data/3D/gm_Ex.data", indStart=indStart, indEnd=None)[:, 0, :, :]
Ex_r_xplane = GetArrays("../data/3D/gr_Ex.data", indStart=indStart, indEnd=None)[:, 0, :, :]
Ex_l_xplane = GetArrays("../data/3D/gl_Ex.data", indStart=indStart, indEnd=None)[:, 0, :, :]
Ex_u_xplane = GetArrays("../data/3D/gu_Ex.data", indStart=indStart, indEnd=None)[:, 0, :, :]
Ex_d_xplane = GetArrays("../data/3D/gd_Ex.data", indStart=indStart, indEnd=None)[:, 0, :, :]
Ex_rr_xplane = GetArrays("../data/3D/grr_Ex.data", indStart=indStart, indEnd=None)[:, 0, :, :]
Ex_ll_xplane = GetArrays("../data/3D/gll_Ex.data", indStart=indStart, indEnd=None)[:, 0, :, :]
Ex_uu_xplane = GetArrays("../data/3D/guu_Ex.data", indStart=indStart, indEnd=None)[:, 0, :, :]
Ex_dd_xplane = GetArrays("../data/3D/gdd_Ex.data", indStart=indStart, indEnd=None)[:, 0, :, :]


Ex_m_yplane = GetArrays("../data/3D/gm_Ey.data", indStart=indStart, indEnd=None)[:, :, 0, :]
Ex_r_yplane = GetArrays("../data/3D/gr_Ey.data", indStart=indStart, indEnd=None)[:, :, 0, :]
Ex_l_yplane = GetArrays("../data/3D/gl_Ey.data", indStart=indStart, indEnd=None)[:, :, 0, :]
Ex_f_yplane = GetArrays("../data/3D/gf_Ey.data", indStart=indStart, indEnd=None)[:, :, 0, :]
Ex_b_yplane = GetArrays("../data/3D/gb_Ey.data", indStart=indStart, indEnd=None)[:, :, 0, :]
Ex_rr_yplane = GetArrays("../data/3D/grr_Ey.data", indStart=indStart, indEnd=None)[:, :, 0, :]
Ex_ll_yplane = GetArrays("../data/3D/gll_Ey.data", indStart=indStart, indEnd=None)[:, :, 0, :]
Ex_ff_yplane = GetArrays("../data/3D/gff_Ey.data", indStart=indStart, indEnd=None)[:, :, 0, :]
Ex_bb_yplane = GetArrays("../data/3D/gbb_Ey.data", indStart=indStart, indEnd=None)[:, :, 0, :]

import pickle
fileParams = open("../data/3D/params.param", "rb")
params = pickle.load(fileParams)
fileParams.close()
print(params)
```

```python
import matplotlib.animation as animation


z_m = np.linspace(params['gm_r0'][2], params['gm_r1'][2], Ex_m_xplane.shape[2])
y_m = np.linspace(params['gm_r0'][1], params['gm_r1'][1], Ex_m_xplane.shape[1])
z_r = np.linspace(params['gr_r0'][2], params['gr_r1'][2], Ex_r_xplane.shape[2])
y_r = np.linspace(params['gr_r0'][1], params['gr_r1'][1], Ex_r_xplane.shape[1])
z_l = np.linspace(params['gl_r0'][2], params['gl_r1'][2], Ex_l_xplane.shape[2])
y_l = np.linspace(params['gl_r0'][1], params['gl_r1'][1], Ex_l_xplane.shape[1])
z_u = np.linspace(params['gu_r0'][2], params['gu_r1'][2], Ex_u_xplane.shape[2])
y_u = np.linspace(params['gu_r0'][1], params['gu_r1'][1], Ex_u_xplane.shape[1])
z_d = np.linspace(params['gd_r0'][2], params['gd_r1'][2], Ex_d_xplane.shape[2])
y_d = np.linspace(params['gd_r0'][1], params['gd_r1'][1], Ex_d_xplane.shape[1])
z_rr = np.linspace(params['grr_r0'][2], params['grr_r1'][2], Ex_rr_xplane.shape[2])
y_rr = np.linspace(params['grr_r0'][1], params['grr_r1'][1], Ex_rr_xplane.shape[1])
z_ll = np.linspace(params['gll_r0'][2], params['gll_r1'][2], Ex_ll_xplane.shape[2])
y_ll = np.linspace(params['gll_r0'][1], params['gll_r1'][1], Ex_ll_xplane.shape[1])
z_uu = np.linspace(params['guu_r0'][2], params['guu_r1'][2], Ex_uu_xplane.shape[2])
y_uu = np.linspace(params['guu_r0'][1], params['guu_r1'][1], Ex_uu_xplane.shape[1])
z_dd = np.linspace(params['gdd_r0'][2], params['gdd_r1'][2], Ex_dd_xplane.shape[2])
y_dd = np.linspace(params['gdd_r0'][1], params['gdd_r1'][1], Ex_dd_xplane.shape[1])


Z_m, Y_m = np.meshgrid(z_m, y_m, indexing="xy")
Z_r, Y_r = np.meshgrid(z_r, y_r, indexing="xy")
Z_l, Y_l = np.meshgrid(z_l, y_l, indexing="xy")
Z_u, Y_u = np.meshgrid(z_u, y_u, indexing="xy")
Z_d, Y_d = np.meshgrid(z_d, y_d, indexing="xy")
Z_rr, Y_rr = np.meshgrid(z_rr, y_rr, indexing="xy")
Z_ll, Y_ll = np.meshgrid(z_ll, y_ll, indexing="xy")
Z_uu, Y_uu = np.meshgrid(z_uu, y_uu, indexing="xy")
Z_dd, Y_dd = np.meshgrid(z_dd, y_dd, indexing="xy")

Dz_m = params['gm_r1'][2] - params['gm_r0'][2]
Dy_m = params['gm_r1'][1] - params['gm_r0'][1]

e_thresh = 1.0
def animate_E(n):
    plt.clf()
    e_max = np.max([np.max(np.abs(Ex_m_xplane[n])), np.max(np.abs(Ex_r_xplane[n])), 
                    np.max(np.abs(Ex_l_xplane[n])), np.max(np.abs(Ex_u_xplane[n])),
                    np.max(np.abs(Ex_d_xplane[n])),
                    np.max(np.abs(Ex_rr_xplane[n])), np.max(np.abs(Ex_ll_xplane[n])),
                    np.max(np.abs(Ex_uu_xplane[n])), np.max(np.abs(Ex_dd_xplane[n]))
                   ])
    if e_max > e_thresh:
        e_max = e_thresh
    fig = plt.pcolormesh(Z_m, Y_m, np.abs(Ex_m_xplane[n, :,:]), cmap="rainbow", vmax=e_max)
    plt.pcolormesh(Z_r, Y_r, np.abs(Ex_r_xplane[n, :,:]), cmap="rainbow", vmax=e_max)
    plt.pcolormesh(Z_l, Y_l, np.abs(Ex_l_xplane[n, :,:]), cmap="rainbow", vmax=e_max)
    plt.pcolormesh(Z_u, Y_u, np.abs(Ex_u_xplane[n, :,:]), cmap="rainbow", vmax=e_max)
    plt.pcolormesh(Z_d, Y_d, np.abs(Ex_d_xplane[n, :,:]), cmap="rainbow", vmax=e_max)
    plt.pcolormesh(Z_rr, Y_rr, np.abs(Ex_rr_xplane[n, :,:]), cmap="rainbow", vmax=e_max)
    plt.pcolormesh(Z_ll, Y_ll, np.abs(Ex_ll_xplane[n, :,:]), cmap="rainbow", vmax=e_max)
    plt.pcolormesh(Z_uu, Y_uu, np.abs(Ex_uu_xplane[n, :,:]), cmap="rainbow", vmax=e_max)
    plt.pcolormesh(Z_dd, Y_dd, np.abs(Ex_dd_xplane[n, :,:]), cmap="rainbow", vmax=e_max)
    plt.colorbar()
    plt.pause(0.05)
    
    return fig

save_animation = False

if save_animation:
    fig = plt.figure(figsize=(7,7*(Dy_m/Dz_m)))
    anim = animation.FuncAnimation(fig, animate_E, frames=Ex_m_xplane.shape[0], interval=1, repeat=False)
    anim.save("../data/3D/" + 'Efield-anim.mp4', writer="ffmpeg", fps=15, dpi=200)
else:    
    plt.ion()
    plt.figure(figsize=(12,12*(Dy_m/Dz_m)))
    for n in range(Ex_m_xplane.shape[0]):
        animate_E(n)

```
