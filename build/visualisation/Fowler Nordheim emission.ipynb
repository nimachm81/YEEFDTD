```python
%pylab inline
import numpy as np


def FowlerNordheimEmissionCurrent(eField_V_cm, workFunction_eV):
    current_c_cm2 = 1.55e-6 * eField_V_cm**2 / workFunction_eV * \
        np.exp(-6.86e7 * workFunction_eV * np.sqrt(workFunction_eV) / eField_V_cm)
    
    return current_c_cm2
    
n_pts = 1000
workFunction_eV = 4.5  ## eV
eField_V_cm = np.linspace(1.0e5, 1.0e8, n_pts)
current_c_cm2 = FowlerNordheimEmissionCurrent(eField_V_cm, workFunction_eV)

plot(eField_V_cm, current_c_cm2)
```
